# Vector と DataFrame の簡単な作り方

このnotebookは https://github.com/SciRuby/sciruby-notebooks/blob/master/Data%20Analysis/Creation%20of%20Vector%20and%20DataFrame.ipynb を実用部だけを抜粋し独自の簡単化を行ったものです。また現状に合わないものは変更を加えています。

```ruby
require 'daru'
samples = 1000
```

```ruby
a = Daru::Vector.new_with_size(samples) {r=rand(5); r==4 ? nil: r}
```

```ruby
b = Daru::Vector.new_with_size(samples) {r=rand(5); r==4 ? nil: r}
```

```ruby
ds = Daru::DataFrame.new({:a=>a,:b=>b}, order: [:b, :a])
```
