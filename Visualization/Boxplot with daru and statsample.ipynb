# daru と statsample を使った boxplot の作り方


```ruby
require 'daru'
require 'statsample'
include Statsample::Shorthand

n = 30
a = rnorm(n-1,50,10)
b = rnorm(n, 30,5)
c = rnorm(n,5,1)
```

```ruby
IRuby.svg boxplot(:vectors=>[a,b,c],:width=>300, :height=>300, :groups=>%w{first first second}, :minimum=>0).to_svg
```
